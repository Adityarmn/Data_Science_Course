```python
# subplot 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows = 2,
                                          ncols = 2,
                                          figsize = (10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(num_butter_prices.keys(),num_butter_prices.values());
ax4.hist(np.random.randn(1000))
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10,5))
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10),np.random.random(10));
ax[1, 0].bar(num_butter_prices.keys(), num_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize =(10,30),subplots=True)
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
heart_disease.plot(kind = "scatter",
                   x = "age",
                   y = "chol",
                   c = "target");
```

```python
# Object-Oriented (OO) combined with pyplot method
fix, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',
        x = 'age',
        y = 'chol',
        c = 'target',
        ax=ax);
ax.set_xlim([45,100]);
```

```python
# OO from scratch
fig, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(
             x = over_50["age"],
             y = over_50["chol"],
             c = over_50["target"])
# Customized the plot
ax.set(title = "Age and Cholesterol comparison",
       xlabel = "Age",
       ylabel = "Cholesterol")

# Add legend
ax.legend(*scatter.legend_elements(),title = "target")

# Add horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = '--')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10),
                              sharex = True)

scatter = ax0.scatter(x = over_50["age"],
                      y = over_50['chol'],
                      c = over_50['target'])

ax0.set(title = "Age dan Cholesterol comparison",
        ylabel = "Cholesterol")

ax0.legend(*scatter.legend_elements(),title ="Target")

ax0.axhline(over_50["chol"].mean(),
           linestyle = '--')
scatter1 = ax1.scatter(x = over_50["age"],
                       y = over_50["thalach"],
                       c = over_50["target"])
ax1.set(title = "Heart rate and Age comparison",
       xlabel = "Age",
       ylabel = "Maximum Heart Rate")

ax1.legend(*scatter1.legend_elements(),title = "target")

ax1.axhline(over_50["thalach"].mean(),
            linestyle = "--")                     
fig.suptitle(t = "Heart Disease Analysis", fontsize = 16, fontweight = 'bold')                     
```
